```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('img/baboon.jpeg')
im_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(im_rgb)
```

```python
R = im_rgb[:, :, 0]
G = im_rgb[:, :, 1]
B = im_rgb[:, :, 2]

plt.imshow(np.hstack([R, G, B]))
```

```python
color = ('b', 'g', 'r')
for i, c in enumerate(color):
    hist = cv2.calcHist([im_rgb], [i], None, [256], [0, 256])
    plt.plot(hist, color=c)
    plt.xlim([0, 256])

plt.xlabel('Rango dinmico')
plt.ylabel('Nmero de pixeles')
plt.show()
```

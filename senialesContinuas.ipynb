```python
def sinc(x):
    if x == 0:
        return 1
    else:
        return np.sin(x)/x


def interpola(x, t, T):
    N = len(x)
    val = 0
    for i in range(-30*N, 30*N):
        n = (i % N)
        if n < 0:
            n = N + n
        val += x[n] * sinc((t - i * T) * np.pi / T)
    return val

```

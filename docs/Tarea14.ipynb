```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def valsPhi(n, k, N):
    return np.exp(-1j*(2*np.pi*n*k/N))
```

## a) Resolviendo el sistema (1/N)Phi X = x,

```python
def ATDF(N, matPhi, vecX):
    return 1/N * np.dot(matPhi, vecX)
```

## b) Haciendo X = (N) Phi^T* x

```python
def TDF(N, matPhi, vecx):
    invPhi = np.linalg.inv(matPhi)
    return N*np.dot(invPhi, vecx)
```

## c) Probar los resultado utilizando la señal x[n] = 4 cos(10 pi n/N) con N = 128.

```python

N = 128

PHI = np.array([[valsPhi(n, k, N) for n in range(N)] for k in range(N)])

x = np.array([ 4*np.cos((10*np.pi*n)/N) for n in range(N)])

X = TDF(N, PHI, x)
# print(X)
# print(ATDF(N, PHI, X))
```

```python
plt.title("Coeficientes de X")
plt.plot(X)
plt.show()
```

```python
plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 1)
plt.title("Señal Original")
plt.plot(x)
plt.grid()

plt.subplot(1, 2, 2)
plt.title("Señal Reconstruida Con Fourier")
n = np.arange(N)
y = 1/N*PHI.dot(X)
plt.plot(n, y)
plt.grid()
```

### Calcular las TDF de un conjunto de señales considerar que cada una de las señales tienen N=128 muestras. Las señales son: 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft
from scipy import stats
```

$$N = 128 $$

```python
N=128
```

## a) Una exponencial compleja con frecuencia 40pi/N

$$ x[n] = e^{j(2\pi/N)k_0 n} $$

$$ x[n] = e^{j(2\pi/128)20 n} $$

$$e^{j(2\pi/128)20 n}    \xrightarrow{\mathscr{F}} 128 \delta(k-20)$$

A continuación podemos ver como la señal presentara un pico en n = 20 con magnitud 128.

```python
x = np.array([np.exp(1j*(40*np.pi*n/N)) for n in range(N)])
plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 1)
plt.title("Señal Original")
plt.plot(x)
plt.grid()

plt.subplot(1, 2, 2)


X = fft(x)
Xreal = X.real
Xim = X.imag
plt.title("FFT Exponencial Compleja")
plt.plot(Xreal, label="Parte Real")
plt.plot(Xim, label="Parte Imaginaria")
plt.legend()
plt.grid()
```

## b) Un seno con frecuencia 20pi/N,

$$ x[n] = sen((2\pi/N)k_0 n) $$

$$ x[n] = sen((20\pi/128)n) = sen((2\pi/128)10 n) $$

$$ sen((2\pi/128)10 n)    \xrightarrow{\mathscr{F}} -64j \delta(k-10) + 64j \delta(k-(128-10)$$
$$-64j \delta(k-10) + 64j \delta(k-118))$$

A continuación podemos ver como la señal presentara un picos en n= 10 y 118 con magnitud -64 y 64 respectivamente.

```python
x = np.array([np.sin((20*np.pi*n)/N) for n in range (N)])


plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
plt.title("Señal Original")
plt.plot(x)
plt.grid()

plt.subplot(1, 2, 2)

X = fft(x)
Xreal = X.real
Xim = X.imag
print(Xim[118])
plt.title("FFT Seno")
plt.plot(Xreal, label="Parte Real")
plt.plot(Xim, label="Parte Imaginaria")
plt.legend()
plt.grid()
plt.show()
```

h) Una gaussiana con media 0 y varianza 10

$$ g[n] = \frac{1}{\sigma \sqrt{2\pi}}e^{-n^2/2\sigma^2} $$

$$ g[n] = \frac{1}{10 \sqrt{2\pi}}e^{-n^2/2*10^2} $$

$$ g[n] = \frac{1}{10 \sqrt{2\pi}}e^{-n^2/200} $$

$$ g[n] = \frac{1}{10 \sqrt{2\pi}}e^{-n^2/200} \xrightarrow{\mathscr{F}} \frac{1}{10 k \sqrt{2\pi}}e^{-k^2/200} $$

Se puede observar que la transformada de Fourier de una gaussiana es otra gaussiana con diferente escalamiento.

```python

#
# Crear gaussiana con media 0 desviación estándar 10
#
mu = 0
std = 10
snd = stats.norm(mu, std)

x = np.linspace(-64, 64, 128)
print(len(x))
G = snd.pdf(x)
x = np.hstack([G[64:], G[:63]])

print(np.sum(x))

plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
plt.title("Señal Original")
plt.plot(x)
plt.grid()

plt.subplot(1, 2, 2)


# plt.plot(x)
plt.title("Gaussiana m=0 varianza =10")
X = fft(x)
Xreal = X.real
Xim = X.imag
plt.plot(Xreal, label="Parte Real")
plt.plot(Xim, label="Parte Imaginaria")
plt.legend()
plt.grid()
```

### Dado una se単al con 128 muestra conformada por un tren de pulsos con periodos T = 2, 4, 8 y 16 determinar la TDF de cada una de las se単ales. En todos los casos graficar la parte real y la parte imaginaria.

```python
import numpy as np
from scipy.ndimage import shift
from scipy.fft import fft, ifft
import matplotlib.pyplot as plt
```

```python
N = 128


def graficarTren(x, T):
    plt.figure(figsize=(20, 10))
    plt.subplot(1, 2, 1)
    plt.title(f'Tren de pulsos T={T}')
    plt.plot(x,'o' ,label=f'Tren de pulsos PR{T}')
    plt.plot(x.imag, label='Tren de Pulsos PI')
    plt.grid()
    plt.legend()

    plt.subplot(1, 2, 2)
    X = fft(x)
    plt.title(f'FFT tren de pulsos T={T}')
    plt.plot(X.real, label='Parte real FFT')
    plt.plot(X.imag, label='Parte imaginaria FFT')
    plt.grid()
    plt.legend()
    plt.show()
```

## Tren de pulsos con periodo T = 2

Se puede observar que la FFT tiene un comportamiento similar a la se単al original, pero con un periodo de 64 muestras, es decir, la mitad de la se単al original. Ademas se puede observar que los picos de la FFT tienen una amplitud de 64 es decir de N/2.

```python
x = np.array([1 if n%2 ==0 else 0 for n in range(N)])
graficarTren(x,2)
```

## Tren de impulsos con periodo T = 4

```python
x = np.array([1 if n%4 ==0 else 0 for n in range(N)])
graficarTren(x,4)
```

## Tren de impulsos con periodo T = 8

```python
x = np.array([1 if n%8 ==0 else 0 for n in range(N)])
graficarTren(x,8)
```

## Tren de impulsos con periodo T = 16

La siguiente gr叩fica tiene un comportamiento peculiar, ya que en este caso la FFT tiene mas picos por lo que utilizar la FFT como una forma de compresi坦n para esta se単al no seria muy eficiente. Se puede tambien observar que la amplitud es de 8 es decir N/16.

```python
x = np.array([1 if n%16 ==0 else 0 for n in range(N)])
graficarTren(x,16)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def valsPhi(n, k, N):
    return np.exp(-1j*(2*np.pi*n*k/N))


def TDF(N, matPhi, vecx):
    invPhi = np.linalg.inv(matPhi)
    return N*np.dot(invPhi, vecx)


def ATDF(N, matPhi, vecX):
    return 1/N * np.dot(matPhi, vecX)
```

```python

N = 128

PHI = np.array([[valsPhi(n, k, N) for n in range(N)] for k in range(N)])

# x = np.array([1, 1, 1, 0, 0, 0, 0, 1, 1])
# x = np.array([1, 4, 1, 1])
# x = np.array([1 if n <= 10 else 1 if n > N-10 else 0 for n in range(N)])
x = np.array([1 if n-20 > 0 else 0 for n in range(N)])
# x = np.array([4*np.cos((10*np.pi*n)/N) for n in range(N)])

print("PHI: \n")
print(PHI)
print("X: \n")
X = TDF(N, PHI, x)
print(X)
print("x: \n")
print(ATDF(N, PHI, X))
```

```python
plt.plot(X.real)
plt.plot(X.imag)
```

```python
plt.figure(figsize=(20, 10))

plt.subplot(1, 2, 1)
plt.title("Seal Original")
plt.plot(x)
plt.grid()

plt.subplot(1, 2, 2)
plt.title("Seal Reconstruida Con Fourier")
n = np.arange(N)
y =1/N*PHI.dot(X)
plt.plot(n,y)
plt.grid()
```

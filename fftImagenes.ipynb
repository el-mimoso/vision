```python
import numpy as np
from scipy.ndimage import shift
from scipy.signal  import convolve2d
from scipy.fft import fft, ifft, fft2
import matplotlib.pyplot as plt
```

```python
N = 100
img = np.array([[np.cos((2*np.pi/100)*(10*x+20*y))
               for y in range(100)] for x in range(100)])

fftimg = fft2(img)

plt.imshow(img, cmap='gray')
plt.show()
plt.imshow(fftimg.real, cmap='gray')
plt.show()
```
